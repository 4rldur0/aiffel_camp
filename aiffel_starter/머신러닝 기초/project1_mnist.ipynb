## 데이터 준비


```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
digit = load_digits(as_frame=True)
```

```python
digit.data.head()
```

```python
digit.target
```

```python
digit.target_names
```

```python
digit.target.value_counts()
```

```python
digit.data.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    digit.data, digit.target, random_state=42
)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 모델 학습
타겟의 분포가 비슷하니 정확도를 평가지표로 삼아도 괜찮을 것이라 판단했습니다  

모든 값이 0~255의 범위 안에 있어 스케일링은 안해도 괜찮다 생각했습니다

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier(random_state=42)
model_tree.fit(X_train, y_train)
y_pred_tree = model_tree.predict(X_test)

print(classification_report(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rfc = RandomForestClassifier(random_state=42)
model_rfc.fit(X_train, y_train)
y_pred_rfc = model_rfc.predict(X_test)

print(classification_report(y_test, y_pred_rfc))
```

```python
from sklearn.svm import SVC

model_svm = SVC(random_state=42)
model_svm.fit(X_train, y_train)
y_pred_svm = model_svm.predict(X_test)

print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.linear_model import SGDClassifier

model_sgd = SGDClassifier(random_state=42)
model_sgd.fit(X_train, y_train)
y_pred_sgd = model_sgd.predict(X_test)

print(classification_report(y_test, y_pred_sgd))
```

```python
from sklearn.linear_model import LogisticRegression

model_logi = LogisticRegression(random_state=42)
model_logi.fit(X_train, y_train)
y_pred_logi = model_logi.predict(X_test)

print(classification_report(y_test, y_pred_logi))
```

```python
from xgboost import XGBClassifier

model_xgb = XGBClassifier(random_state=42)
model_xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict(X_test)

print(classification_report(y_test, y_pred_xgb))
```

## XGB 파인튜닝

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
from xgboost import XGBClassifier

params = {
    "n_estimators": list(range(100, 500)),
    "learning_rate": list(np.logspace(np.log10(0.0001), np.log10(0.1))),
    "max_depth": list(np.arange(3, 10)),
    # "reg_alpha": list(np.linspace(0, 1)),
    # "reg_lambda": list(np.linspace(0, 1)),
    "gamma": list(np.logspace(np.log10(0.01), np.log10(20))),
    "colsample_bytree": list(np.linspace(0.6, 1, 10)),
    "subsample": list(np.linspace(0.5, 1, 10)),
}

model_xgb = XGBClassifier(random_state=42, n_jobs=-1)
tuner = RandomizedSearchCV(model_xgb, params, cv=3, n_iter=50)
tuner.fit(X_train, y_train)
```

```python
tuner.best_params_
```

```python
model_fin = XGBClassifier(
    random_state=42,
    subsample=0.7777777777777778,
    n_estimators=494,
    max_depth=9,
    learning_rate=0.028117686979742307,
    gamma=0.040393513624019946,
    colsample_bytree=0.8666666666666667,
)
model_fin.fit(X_train, y_train)
y_pred_fin = model_fin.predict(X_test)

print(classification_report(y_test, y_pred_fin))
```
